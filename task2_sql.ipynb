```python
# Import required libraries
# Pandas is used for data manipulation
# SQLite3 is used to create and connect to the database

import pandas as pd
import sqlite3
```

```python
# Load the CSV dataset into a Pandas DataFrame
# Replace the file path with your dataset location

df = pd.read_csv(r"C:\Users\somes\OneDrive\Pictures\Attachments\Desktop\Documents\apex planet\task-2\SuperMarket Analysis.csv")

# Display the first five rows
df.head()
```

```python
# Display dataset information
# Shows number of rows, columns, data types and missing values

df.info()
```

```python
# Display all column names
# Helps while writing SQL queries

print(df.columns.tolist())
```

```python
# Create (or connect to) SQLite database
# A new database file named supermarket.db will be created

conn = sqlite3.connect("supermarket.db")

print("Database Connected Successfully")
```

```python
# Store the DataFrame into SQLite
# Creates a table named 'sales'

df.to_sql("sales", conn, if_exists="replace", index=False)

print("Table Created Successfully")
```

```python
# Display first 10 rows from the sales table

query = """
SELECT *
FROM sales
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
# Calculate total sales

query = """
SELECT SUM(Sales) AS Total_Sales
FROM sales;
"""

pd.read_sql(query, conn)
```

```python
# Calculate average sales

query = """
SELECT AVG(Sales) AS Average_Sales
FROM sales;
"""

pd.read_sql(query, conn)
```

```python
# Display top 10 highest sales transactions

query = """
SELECT *
FROM sales
ORDER BY Sales DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
# Calculate total sales for each city

query = """
SELECT City,
SUM(Sales) AS TotalSales
FROM sales
GROUP BY City;
"""

pd.read_sql(query, conn)
```

```python
# Count customers based on gender

query = """
SELECT Gender,
COUNT(*) AS Customers
FROM sales
GROUP BY Gender;
"""

pd.read_sql(query, conn)
```

```python
# Calculate total sales for each product line

query = """
SELECT [Product line],
SUM(Sales) AS TotalSales
FROM sales
GROUP BY [Product line]
ORDER BY TotalSales DESC;
"""

pd.read_sql(query, conn)
```

```python
# Count the number of transactions for each payment method

query = """
SELECT Payment,
COUNT(*) AS TotalTransactions
FROM sales
GROUP BY Payment;
"""

pd.read_sql(query, conn)
```

```python
# Calculate monthly sales

query = """
SELECT substr(Date,1,7) AS Month,
SUM(Sales) AS MonthlySales
FROM sales
GROUP BY Month;
"""

pd.read_sql(query, conn)
```

```python
# Calculate total sales based on customer type

query = """
SELECT [Customer type],
SUM(Sales) AS TotalSpent
FROM sales
GROUP BY [Customer type]
ORDER BY TotalSpent DESC;
"""

pd.read_sql(query, conn)
```

```python
# Calculate average customer rating for each city

query = """
SELECT City,
AVG(Rating) AS AverageRating
FROM sales
GROUP BY City;
"""

pd.read_sql(query, conn)
```

```python
# Find the highest sale amount

query = """
SELECT MAX(Sales) AS HighestSale
FROM sales;
"""

pd.read_sql(query, conn)
```

```python
# Find the lowest sale amount

query = """
SELECT MIN(Sales) AS LowestSale
FROM sales;
"""

pd.read_sql(query, conn)
```

```python
# Count transactions in each branch

query = """
SELECT Branch,
COUNT(*) AS TotalTransactions
FROM sales
GROUP BY Branch;
"""

pd.read_sql(query, conn)
```

```python
# Close the SQLite database connection

conn.close()

print("Database Connection Closed Successfully")
```
